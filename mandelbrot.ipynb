```python
import torch as th
import matplotlib.pyplot as plt
```

```python
def mBrot(nIm,nIter):
    p=0
    for norm in range(nIter):
        if abs(p)>2:
            return norm
        p= p * p + nIm
    
    return nIter
```

```python
def mBrotSet(limits,box,nIter,device):
    x=th.linspace(limits[0],limits[1],box[0],device=device)
    y=th.linspace(limits[2],limits[3],box[1],device=device)
    mBSet=th.zeros(box[0],box[1],device=device)
    
    for i in range(box[1]):
        for j in range(box[0]):
            nIm=th.complex(x[j],y[i])
            mBSet[i,j]=mBrot(nIm,nIter)
    
    return mBSet
```

```python
limits=[-2.,1.,-1.5,1.5]
box=[1000,1000]
nIter=10
device="cpu"
if th.cuda.is_available():
    device="cuda"
```

```python
mBrotImage=mBrotSet(limits,box,nIter,device)
```

```python

plt.imshow(mBrotImage.cpu().numpy(),extent=limits,cmap='inferno')
plt.colorbar()
```

```python
mBrotImage=mBrotSet(limits,box,nIter*10,device)
plt.imshow(mBrotImage.cpu().numpy(),extent=limits,cmap='inferno')
plt.colorbar()
```

```python
mBrotImage=mBrotSet(limits,box,nIter*100,device)
plt.imshow(mBrotImage.cpu().numpy(),extent=limits,cmap='inferno')
plt.colorbar()
```
